```julia
using CSV, Feather, DataFrames, Dates
```

```julia
df = CSV.File("NFL Play by Play 2009-2018 (v5).csv") |> DataFrame
```

```julia
print(names(df))
```

```julia
cols_list = [:posteam, :yardline_100, 
    :quarter_seconds_remaining, :half_seconds_remaining, :game_seconds_remaining, 
    :drive, :qtr, :down, :ydstogo, :desc, :shotgun, :no_huddle,
    :posteam_timeouts_remaining, :defteam_timeouts_remaining, :posteam_score, :defteam_score,
    :score_differential, :rush_attempt, :pass_attempt, :season]
```

```julia
filter!(row -> row[:rush_attempt] != row[:pass_attempt], df)
```

```julia
number_cols = [:quarter_seconds_remaining, :posteam_timeouts_remaining, :defteam_timeouts_remaining, 
    :posteam_score, :defteam_score, 
    :score_differential]
for col in number_cols
    df[col] = parse.(Int64, df[col])
end
```

```julia
filter!(row -> row[:down] != "NA", df)
df[:down] = parse.(Int8, df[:down])
```

```julia
df[:rush_attempt] = parse.(Int8, df[:rush_attempt])
df[:pass_attempt] = parse.(Int8, df[:pass_attempt])
```

```julia
for col in [:half_seconds_remaining, :game_seconds_remaining]
    filter!(row -> row[col] != "NA", df)
    df[col] = replace.(df[col], "e3" .=> "000")
    df[col] = parse.(Int64, df[col])
end
```

```julia
filter(row -> Dates.month(row) < 8, df[:game_date])
```

```julia
function parse_season(d::Date)
    if Dates.month(d) > 8
        return Dates.year(d)
    else
        return Dates.year(d) - 1
    end
end
```

```julia
df[:season] = parse_season.(df[:game_date])
```

```julia
Feather.write("season_data.feather", df[cols_list])
```
